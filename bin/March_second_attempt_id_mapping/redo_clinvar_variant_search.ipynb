```python
testing_df = df_meta.loc[[4, 5, 3251244,3251246]+[3211451,3211449, 3211581, 825282, 328697, 15383, 16017, 12772, 46304, 47999, 552269]]
testing_df
```

```python
# showing wield input regex have to extract
testing_df = testing_df[['Name', 'GeneSymbol', 'Assembly', 'ChromosomeAccession','ReferenceAlleleVCF','AlternateAlleleVCF' ]]
testing_df['CDS_position'] = testing_df['Name'].str.extract(r'c\.\*?(\d+)')
testing_df['AlleleChange'] = testing_df['Name'].str.extract(r'c\.\d+([A-Z])>') + '/' + testing_df['Name'].str.extract(r'>([A-Z])')
# put na for those entires of intron changes
import numpy as np
testing_df['Protein_Patterns'] = testing_df['Name'].str.extract(r'p\.(\w+\d+[A-Z][a-z]+)', expand=False).fillna(np.nan)
testing_df
```

```python
# need to consider conditions of:
# p.Tyr5033=  synonymous mutation
# p.Ter5039*  premature stop codon

def extract_amino_acid(row, before = False, pos = False, after = False):
    import re
    try:
        if before:
            first_aa = row[:3]
            return first_aa
        if pos:
            pos = int(re.search(r'\d+', row).group())
            return pos
        # the following command extract the index within the string where the position number ends
        if after:
            first_aa = row[:3]
            position_end_index = re.search(r'\d+', row).span()[-1]
            later_part = row[position_end_index:]
            if later_part == '=':
                return first_aa
            elif later_part == '*':
                return 'STOP'
            else:
                return later_part
    except TypeError:
        import numpy as np
        return np.nan

testing_df['Amino_acids_before_3letters'] = testing_df['Protein_Patterns'].apply(extract_amino_acid, before = True)
testing_df['Protein_positions'] = testing_df['Protein_Patterns'].apply(extract_amino_acid, pos = True)
testing_df['Amino_acids_after_3letters'] = testing_df['Protein_Patterns'].apply(extract_amino_acid, after = True)
testing_df['Protein_positions'] = testing_df['Protein_positions'].astype('Int64')
testing_df
```

```python
# drop introns and premature stop codon
testing_df = testing_df[(testing_df['Protein_Patterns'].notnull()) &
                        (testing_df['Amino_acids_before_3letters'] != 'Ter') &
                        (testing_df['Amino_acids_after_3letters'] != 'Ter') &
                        (testing_df['AlleleChange'].notnull())]
testing_df
```
