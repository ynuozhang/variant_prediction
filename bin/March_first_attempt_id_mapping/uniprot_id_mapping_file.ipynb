```python
import pandas as pd
```

```python
df = pd.read_csv('./data/metas/HUMAN_9606_idmapping.dat', sep='\t', header=None)
df.columns = ['UniProtKB-AC', 'ID_type', 'ID']
df
```

```python
df[df['UniProtKB-AC']=='Q6ZMK1']
```

```python
# get rows that only related to Gene Names
df_genename = df[df['ID_type'].str.contains('UniProtKB-ID', regex=True)]
df_genename.drop(['ID_type'], axis=1, inplace=True)
df_genename.drop_duplicates(inplace=True)
df_genename
```

```python
df_genename[df_genename.duplicated(subset=['UniProtKB-AC'], keep=False)].sort_values('UniProtKB-AC')
```

```python
df_genename.to_csv('./data/metas/uniprotid_uniprot_mapping.csv', index=False)
```

```python
df[df['UniProtKB-AC'] == 'A4Z6T7']
```

```python
# get rows that only related and UniParc
df_uniparc = df[df['ID_type'].str.contains('UniParc', regex=True)]
df_uniparc.drop(['ID_type'], axis=1, inplace=True)
df_uniparc.drop_duplicates(inplace=True)
df_uniparc
```

```python
df_uniparc.to_csv('./data/metas/unipac_uniprot_mapping.csv', index=False)
```

```python
# remove duplicated entries for canonical proteins (don't use for RefSeq Mapping)
#df_uniparc_unique = df_uniparc.sort_values(by='UniProtKB-AC', ascending=False)
df_uniparc_unique = df_uniparc.sort_values(by='UniProtKB-AC')
#df_grouped = df.groupby('C').agg({'A': lambda x: ','.join(set(x))}).reset_index()
df_uniparc_unique = df_uniparc_unique.drop_duplicates(subset=['ID'])
df_uniparc_unique
```

```python
df_uniparc_unique.to_csv('./data/metas/unipac_uniprot_rmdup.csv', index=False)
```

```python
df_uniparc_unique['UniProtKB-AC'].values.tolist()
```

```python
# get rows that only related to NM and NP, and UniParc
#df_subset = df[(df['ID_type'].str.contains('RefSeq_NT|RefSeq|GeneID', regex=True)) & (df['UniProtKB-AC'].isin(df_grouped['UniProtKB-AC'].values.tolist()))] # this is for canonical extraction
df_subset = df[(df['ID_type'].str.contains('RefSeq_NT|RefSeq|GeneID|UniParc', regex=True))]
df_subset
```

```python
# separate files. GI mapping
df_GI = df_subset[df_subset['ID_type']=='GeneID']
df_GI.drop(['ID_type'], axis=1, inplace=True)
df_GI.drop_duplicates(inplace=True)
df_GI
```

```python
df_GI.to_csv('./data/metas/GI_UniProt_mapping.csv', index=False)
```

```python
df_ref = df_subset[(df_subset['ID_type'] == 'RefSeq')]
#df_ref.drop(['ID_type'], axis=1, inplace=True)
df_ref.drop_duplicates(inplace=True)
df_ref
```

```python
# Check if NM and NP are one to one
df_ref_nm = df_subset[(df_subset['ID_type'] == 'RefSeq_NT')]
#df_ref.drop(['ID_type'], axis=1, inplace=True)
df_ref_nm.drop_duplicates(inplace=True)
df_ref_nm
```

```python
df_ref.to_csv('./data/metas/RefSeq_UniProt_mapping.csv', index=False)
```

```python
df_uniprotkb_uniparc = pd.merge(df_ref, df_uniparc, how='left', on='UniProtKB-AC')
df_uniprotkb_uniparc
```

```python
# there shouldn't be NAs in the merging
df_uniprotkb_uniparc.isna().values.any()
```

```python
# map uniprotkb to unique entries, merge canonical ones
df_rmdup_uniprotkb_uniparc = pd.merge(df_uniprotkb_uniparc, df_uniparc_unique, how='left', left_on='ID_y', right_on='ID')
df_rmdup_uniprotkb_uniparc
```

```python
# there shouldn't be NAs in the merging
df_rmdup_uniprotkb_uniparc.isna().values.any()
```

```python
df_ref_new = df_rmdup_uniprotkb_uniparc[['UniProtKB-AC_y', 'ID_x']]
df_ref_new.columns = ['UniProtKB-AC', 'ID']
df_ref_new.drop_duplicates(inplace=True)
df_ref_new['UniProtKB-AC'].unique()
```

```python
df_ref_new.to_csv('./data/metas/RefSeq_UniProt_mapping_new.csv', index=False)
```

```python
# merge canonical sequences (if they share the same UniParc ID)
merged_df = pd.DataFrame(columns=['UniProtKB-AC', 'ID_type', 'ID'])
for acc in df_subset['UniProtKB-AC'].unique():
    df_subset_ = df_subset.loc[df_subset['UniProtKB-AC'].str.startswith(acc.split('-')[0])]
    if 'UniParc' in df_subset_['ID_type'].values and len(df_subset_['ID'].unique()) == 1:
        merged_df = merged_df.append({'UniProtKB-AC': acc.split('-')[0], 'ID_type': 'UniParc', 'ID': df_subset_['ID'].unique()[0]}, ignore_index=True)
    else:
        merged_df = merged_df.append(df_subset_, ignore_index=True)

# Print the merged dataframe
print(merged_df)
```
