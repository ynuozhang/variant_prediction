```python
"""This script find overlappings between EVE published variant scores and clinvar annotated variants"""
import pandas as pd
import numpy as np
```

```python
# filter clinvar dataset for one stars
df_clinvar = pd.read_csv('./data/metas/final_clinvar_variant_uniprot_id_mapping.csv')
df_clinvar
```

```python
df_uniprotid_uniprotkb_mappings = pd.read_csv('./data/metas/clinvar_variant_uniprotid_uniprotkb_mapping.csv')
df_eve_covered_uniprotkbs = pd.merge(df_clinvar, df_uniprotid_uniprotkb_mappings, how='left', on='UniProtKB-AC')
df_eve_covered_uniprotkbs.dropna(subset=['UniProtKB-ID'], inplace=True)
df_eve_covered_uniprotkbs
```

```python
# double check numbers of unique entries
print(len(df_eve_covered_uniprotkbs['UniProtKB-AC'].unique()))
print(len(df_eve_covered_uniprotkbs['GeneSymbol'].unique()))
print(len(df_eve_covered_uniprotkbs['GeneID'].unique()))
```

```python
other_descriptions = set()
def assign_stars(row):
    if 'criteria provided, single submitter' in row or 'criteria provided, conflicting interpretations' in row:
        return 1
    elif 'criteria provided, multiple submitters' in row:
        return 2
    elif 'expert panel' in row:
        return 3
    elif 'guideline' in row:
        return 4
    elif 'no assertion' in row:
        return None
    else:
        other_descriptions.add(row)
        return None
df_eve_covered_uniprotkbs['stars'] = df_eve_covered_uniprotkbs['ReviewStatus'].apply(assign_stars)
# pandas can handle NaNs as integers now
df_eve_covered_uniprotkbs['stars'] = df_eve_covered_uniprotkbs['stars'].astype('Int64')
df_eve_covered_uniprotkbs = df_eve_covered_uniprotkbs[df_eve_covered_uniprotkbs['stars'] >= 1]
df_eve_covered_uniprotkbs.drop(['ReviewStatus', 'protein_id'], axis=1, inplace=True)
print(other_descriptions)
df_eve_covered_uniprotkbs
```

```python
# double check numbers of unique entries
print(len(df_eve_covered_uniprotkbs['UniProtKB-AC'].unique()))
print(len(df_eve_covered_uniprotkbs['GeneSymbol'].unique()))
print(len(df_eve_covered_uniprotkbs['GeneID'].unique()))
```

```python
# Reassign class
def assign_class(row):
    if 'Benign/Likely benign' in row:
        return 'Benign'
    elif 'Pathogenic/Likely pathogenic' in row:
        return 'Pathogenic'
    elif 'Benign' in row:
        return 'Benign'
    elif 'Likely benign' in row:
        return 'Benign'
    elif 'Pathogenic' in row:
        return 'Pathogenic'
    elif 'Likely pathogenic' in row:
        return 'Pathogenic'
    else:
        raise KeyError("You don't belong here")
variant_selected['classes'] = variant_selected['ClinicalSignificance'].apply(assign_class)
# pandas can handle NaNs as integers now
variant_selected.drop(['Assembly'], axis=1, inplace=True)
#variant_selected = variant_selected[variant_selected['classes'].notnull()]
variant_selected
```
