```python
import pandas as pd
```

```python
path = '/data/rozen/home/e0833634/variant_prediction/data/ucsf_download/content/ALL_hum_isoforms_ESM1b_LLR/'
```

```python
clinvar_meta = pd.read_csv('./7049_preliminary_check.csv', index_col=0)
clinvar_meta.reset_index(inplace=True)
clinvar_meta.drop(['index'], axis=1, inplace=True)
clinvar_meta.dropna(inplace=True)
clinvar_meta
```

```python
clinvar_meta = clinvar_meta.loc[1:3]
clinvar_meta
```

```python
df = pd.read_csv(path+'Q9ULP9_LLR.csv', index_col=0)
df
```

```python
variant_meta = {}
for index, row in clinvar_meta.iterrows():
    r, c = row['Amino_acids_after'], row['Amino_acids_before']+' '+str(row['Protein_positions'])
    try:
        df = pd.read_csv(path+row['uniprot_acc']+'_LLR.csv', index_col=0)
        variant_meta[index] = {'uniprot_acc' : row['uniprot_acc'],
                            'scores': df.loc[r][c],
                            'variant': c+'_'+r,
                               'category': row['classes']
                            }
    except (KeyError, FileNotFoundError):
        pass
variant_meta
```

```python
df_var = pd.DataFrame.from_dict(variant_meta, orient='index')
df_var
```

```python
import matplotlib.pyplot as plt
colors = df_var['category'].unique()
colors
```

```python
import numpy as np
colors = np.array(['red', 'red', 'blue', 'blue'])
```

```python
def assign_binary_class(row):
    if 'Benign/Likely benign' in row or 'Pathogenic/Likely pathogenic' in row:
        return None
    elif 'Benign' in row:
        return 'Benign'
    elif 'Likely benign' in row:
        return 'Benign'
    elif 'Pathogenic' in row:
        return 'Pathogenic'
    elif 'Likely pathogenic' in row:
        return 'Pathogenic'
    else:
        return None
def assign_color(row):
    if 'Benign' in row:
        return 'blue'
    elif 'Pathogenic' in row:
        return 'red'
df_var['colors'] = df_var['binary_class'].apply(assign_color)
df_var
```

```python
df_var['occupy'] = pd.Series([0] * len(df_var))
df_var
```

```python
plt.scatter(df_var['binary_class'], df_var['scores'], c=df_var['colors'])
plt.show()
```

```python
plt.scatter(df_var['occupy'], df_var['scores'])
plt.show()
```

```python
# read previously calculated scores for the known variants
df_s = pd.read_csv('/data/rozen/home/e0833634/variant_prediction/single_VUS/known_variants_scores_organized.csv', index_col=0)
df_s = df_s[df_s['models'] == 'esm1b']
def assign_gaussian_threshold(row):
    if row >= -7.5:
        return 'benign'
    else:
        return 'pathogenic'


df_s['UCSF_hard_threshold'] = df_s['scores'].apply(assign_gaussian_threshold)
df_s
```
