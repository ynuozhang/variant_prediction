```python
df = df[(df['Type'] == 'single nucleotide variant')]
df
```

```python
df_meta = df.copy()
```

```python
def assign_stars(row):
    if 'criteria provided, single submitter' in row or 'criteria provided, conflicting interpretations' in row:
        return 1
    elif 'criteria provided, multiple submitters' in row:
        return 2
    elif 'expert panel' in row:
        return 3
    elif 'guideline' in row:
        return 4
    elif 'no assertion' in row:
        return None
    else:
        #print(row)
        # it's phrased as 'no interpretation for the single variant'
        print('something is wrong')

df_meta['stars'] = df_meta['ReviewStatus'].apply(assign_stars)
# pandas can handle NaNs as integers now
df_meta['stars'] = df_meta['stars'].astype('Int64')
df_meta = df_meta[df_meta['stars'] >= 1]
df_meta.drop(['ReviewStatus'], axis=1, inplace=True)
df_meta
```

```python
df_meta['ClinicalSignificance'].unique()
```

```python
def assign_class(row):
    if 'Benign/Likely benign' in row or 'Pathogenic/Likely pathogenic' in row:
        return None
    elif 'Benign' in row:
        return 'Benign'
    elif 'Likely benign' in row:
        return 'Likely benign'
    elif 'Pathogenic' in row:
        return 'Pathogenic'
    elif 'Likely pathogenic' in row:
        return 'Likely pathogenic'
    else:
        return None

df_meta['classes'] = df_meta['ClinicalSignificance'].apply(assign_class)
# pandas can handle NaNs as integers now
df_meta = df_meta[df_meta['classes'].notnull()]
df_meta
```

```python
# identify the amino acid changes
import numpy as np
df_meta['CDS_position'] = df_meta['Name'].str.extract(r'c\.(\d+)')
df_meta['AlleleChange'] = df_meta['Name'].str.extract(r'c\.\d+([A-Z])>') + '/' + df_meta['Name'].str.extract(r'>([A-Z])')
# put na for those entires of intron changes
df_meta['Protein_Patterns'] = df_meta['Name'].str.extract(r'p\.(\w+\**=*)', expand=False).fillna(np.nan)
df_meta
```

```python
# need to consider conditions of:
# p.Tyr5033=  synonymous mutation
# p.Ter5039*  premature stop codon

def extract_amino_acid(row, before = False, pos = False, after = False):
    import re
    try:
        if before:
            first_aa = row[:3]
            return first_aa
        if pos:
            pos = int(re.search(r'\d+', row).group())
            return pos
        # the following command extract the index within the string where the position number ends
        if after:
            first_aa = row[:3]
            position_end_index = re.search(r'\d+', row).span()[-1]
            later_part = row[position_end_index:]
            if later_part == '=':
                return first_aa
            elif later_part == '*':
                return 'STOP'
            else:
                return later_part
    except TypeError:
        import numpy as np
        return np.nan

df_meta['Amino_acids_before_3letters'] = df_meta['Protein_Patterns'].apply(extract_amino_acid, before = True)
df_meta['Protein_positions'] = df_meta['Protein_Patterns'].apply(extract_amino_acid, pos = True)
df_meta['Amino_acids_after_3letters'] = df_meta['Protein_Patterns'].apply(extract_amino_acid, after = True)
df_meta['Protein_positions'] = df_meta['Protein_positions'].astype('Int64')
df_meta
```

```python
# drop introns and premature stop codon
df_meta = df_meta[(df_meta['Protein_Patterns'].notnull()) & (df_meta['Amino_acids_after_3letters'] != 'Ter')]
df_meta
```

```python
# keep only nonsynonymous changes
amino_acids_dict = {'Ala': 'A', 'Arg': 'R', 'Asn': 'N', 'Asp': 'D', 'Cys': 'C', 'Glu': 'E', 'Gln': 'Q', 'Gly': 'G', 'His': 'H', 'Ile': 'I', 'Leu': 'L', 'Lys': 'K', 'Met': 'M', 'Phe': 'F', 'Pro': 'P', 'Ser': 'S', 'Thr': 'T', 'Trp': 'W', 'Tyr': 'Y', 'Val': 'V'}
df_meta['Amino_acids_before'] = df_meta['Amino_acids_before_3letters'].map(amino_acids_dict)
df_meta['Amino_acids_after'] = df_meta['Amino_acids_after_3letters'].map(amino_acids_dict)
df_meta = df_meta[df_meta['Amino_acids_before'] != df_meta['Amino_acids_after']]
df_meta['Amino_acids'] = df_meta['Amino_acids_before'] + '/' + df_meta['Amino_acids_after']
#df_subset.drop(['Amino_acids_after_3letters', 'Amino_acids_before_3letters', 'Amino_acids_before', 'Amino_acids_after'], axis=1, inplace=True)
df_meta
```

```python
# processing uniprotkb variant metadata into a dictionary
import itertools
import re
uniprot_variant_dict = {}
len_of_disease_entries = 0
with open('./data/metas/humsavar.txt') as f:
    for line in itertools.islice(f, 44, 81966):
        # check weekly record
        # the gene_name seems never exceed 10 character, use ' ' cannot strip between gene name and accession number
        gene_name = line[:10].strip()
        fields = re.split(r'\s+', line[10:].strip())
        try:
            key = fields[0]+'/'+fields[1]+'/'+gene_name
            if key in uniprot_variant_dict.keys():
                # if this critera is hit, this variant has multiple disease OMIM
                uniprot_variant_dict[key]['Disease_name'].append(' '.join(fields[5:]))
                len_of_disease_entries += 1
                #print(uniprot_variant_dict[key], line)
                pass
            else:
                uniprot_variant_dict[key] = {}
                uniprot_variant_dict[key]['Gene_name'] = gene_name
                uniprot_variant_dict[key]['Accession'] = fields[0]
                uniprot_variant_dict[key]['Variant_FTId'] = fields[1]
                uniprot_variant_dict[key]['Amino_acid_change'] = fields[2]
                uniprot_variant_dict[key]['Variant_category'] = fields[3]
                uniprot_variant_dict[key]['dbSNP'] = fields[4]
                uniprot_variant_dict[key]['Disease_name'] = [' '.join(fields[5:])]
        except IndexError:
            print(line)
uniprot_variant_dict = {k: {**v, 'Disease_name':'; '.join(v['Disease_name'])} for k, v in uniprot_variant_dict.items()}
```

```python
# make the above info into dataframe
df_var = pd.DataFrame.from_dict(uniprot_variant_dict, orient='index')
df_var
```

```python
# need to consider conditions of:
# p.Tyr5033=  synonymous mutation
# p.Ter5039*  premature stop codon

def extract_amino_acid(row, before = False, pos = False, after = False):
    import re
    try:
        if before:
            first_aa = row[2:5]
            return first_aa
        if pos:
            pos = int(re.search(r'\d+', row).group())
            return pos
        # the following command extract the index within the string where the position number ends
        if after:
            first_aa = row[2:5]
            position_end_index = re.search(r'\d+', row).span()[-1]
            later_part = row[position_end_index:]
            if later_part == '=':
                return first_aa
            elif later_part == '*':
                return 'STOP'
            else:
                return later_part
    except TypeError:
        import numpy as np
        return np.nan

df_var['Amino_acids_before_3letters'] = df_var['Amino_acid_change'].apply(extract_amino_acid, before = True)
df_var['Protein_positions'] = df_var['Amino_acid_change'].apply(extract_amino_acid, pos = True)
df_var['Amino_acids_after_3letters'] = df_var['Amino_acid_change'].apply(extract_amino_acid, after = True)
df_var['Protein_positions'] = df_var['Protein_positions'].astype('Int64')
# drop introns and premature stop codon
df_var = df_var[(df_var['Amino_acid_change'].notnull()) & (df_var['Amino_acids_after_3letters'] != 'Ter')]
# no premature termination
# keep only nonsynonymous changes
amino_acids_dict = {'Ala': 'A', 'Arg': 'R', 'Asn': 'N', 'Asp': 'D', 'Cys': 'C', 'Glu': 'E', 'Gln': 'Q', 'Gly': 'G', 'His': 'H', 'Ile': 'I', 'Leu': 'L', 'Lys': 'K', 'Met': 'M', 'Phe': 'F', 'Pro': 'P', 'Ser': 'S', 'Thr': 'T', 'Trp': 'W', 'Tyr': 'Y', 'Val': 'V'}
df_var['Amino_acids_before'] = df_var['Amino_acids_before_3letters'].map(amino_acids_dict)
df_var['Amino_acids_after'] = df_var['Amino_acids_after_3letters'].map(amino_acids_dict)
df_var = df_var[df_var['Amino_acids_before'] != df_var['Amino_acids_after']]
df_var['Amino_acids'] = df_var['Amino_acids_before'] + '/' + df_var['Amino_acids_after']
df_var
```
